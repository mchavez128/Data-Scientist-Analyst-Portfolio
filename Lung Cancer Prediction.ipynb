# 2. Modeling

```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing

X = df.drop(["Level_High","Level_Low","Level_Medium"], axis=1)
y_high = df["Level_High"]
y_low = df["Level_Low"]
y_medium = df["Level_Medium"]

X_train, X_test, y_train, y_test = train_test_split(X, y_high, test_size=0.30, random_state=42)
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y_low, test_size=0.30, random_state=42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y_medium, test_size=0.30, random_state=42)
```

```python
X_train.shape, X_train1.shape, X_train2.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

model_pipeline = []
model_pipeline.append(LogisticRegression(solver='liblinear'))
model_pipeline.append(SVC())
model_pipeline.append(KNeighborsClassifier())
model_pipeline.append(DecisionTreeClassifier())
model_pipeline.append(RandomForestClassifier())
model_pipeline.append(GaussianNB())
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix

model_list =['Logistic Regression','SVM', 'KNN', 'Decision Tree',
            'Random Forest', 'Naive Bayes']
acc_list, acc_list1, acc_list2 = [], [], []
auc_list, auc_list1, auc_list2 = [], [], []

for model in model_pipeline:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc_list.append(round(metrics.accuracy_score(y_test, y_pred),3))
    fpr, tpr, _thresholds = metrics.roc_curve(y_test, y_pred)
    auc_list.append(round(metrics.auc(fpr, tpr),3))
```

```python
model_result_high = pd.DataFrame({'Model': model_list, 'Accuracy': acc_list,
                                 'AUC': auc_list}).set_index('Model')*100
model_result_high
```

```python
for model1 in model_pipeline:
    model1.fit(X_train1, y_train1)
    y_pred1 = model1.predict(X_test1)
    acc_list1.append(round(metrics.accuracy_score(y_test1, y_pred1),3))
    fpr, tpr, _thresholds = metrics.roc_curve(y_test1, y_pred1)
    auc_list1.append(round(metrics.auc(fpr, tpr),3))
```
